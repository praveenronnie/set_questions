```python
#1. Why does this feature selection with chi2 fail?
from sklearn.feature_selection import SelectKBest, chi2

X = [[-1, 2, 3], [4, 0, -6], [7, 8, 9], [1, 2, 3]]
y = [0, 1, 0, 1]

selector = SelectKBest(score_func=chi2, k=2)
X_new = selector.fit_transform(X, y)
```

```python
# current issues => presence of negative values
#   chi2 requires non negative values
#
# Solution:
#   scale the features from 0 to 1.

from sklearn.feature_selection import SelectKBest, chi2
from sklearn.preprocessing import MinMaxScaler

X = [[-1, 2, 3], [4, 0, -6], [7, 8, 9], [1, 2, 3]]
y = [0, 1, 0, 1]

scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(X)
selector = SelectKBest(score_func=chi2, k=2)
X_new = selector.fit_transform(x_scaled, y)
print(X_new)
```

```python
#2.
from sklearn.metrics.pairwise import cosine_similarity

# Error: strings instead of numerical features
user1 = ['action', 'comedy', 'romance']
user2 = ['action', 'thriller', 'comedy']

similarity = cosine_similarity([user1], [user2])
```

```python
# current issues => similarity requires numeric but categorical data given
# solution:
#       encode the data

from sklearn.preprocessing import MultiLabelBinarizer

encoded = MultiLabelBinarizer().fit_transform([user1, user2])
similarity = cosine_similarity([encoded[0]], [encoded[1]])
print(similarity)
```

```python
# 6.

from sklearn.metrics.pairwise import cosine_similarity

items = ["action movie", "romantic comedy", "sci-fi thriller"]
similarity = cosine_similarity([items[0]], [items[1]])
```

```python
# current issues => texts are passed as argument for cosine similarity
# solution:
#       * encode the categorical data. In this case the data is not just categorical, it is a text.
#       * encoding for texts --> TF-IDF Vectorizer (Term Frequency Inverse Document Frequency)
#       * converts text to vectors
#       * TF (Term Frequency) → how often a word appears in a document
#       * IDF (Inverse Document Frequency) → how rare or important that word is across all documents


from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

items = ["action movie", "romantic comedy", "sci-fi thriller"]

vectorizer = TfidfVectorizer()
items_vectorized = vectorizer.fit_transform(items)
print(items_vectorized)

# similarity between 'action movie' & 'romantic comedy'
similarity = cosine_similarity(items_vectorized[:1], items_vectorized[1:2])
print("similarity\n", similarity)
```

```python
# current issues -> chi2 accepts only non negative values

from scipy.sparse import csr_matrix
from sklearn.feature_selection import SelectKBest, chi2

X = csr_matrix([[0, -1, 3], [0, 2, 1], [3, -1, 0]])
y = [0, 1, 0]

#! MinMax scaler doesnt support sparse matrix (CSR).
#! MaxAbsScaler support CSR but contains non negative values (-1,1)

# Hence clipping negative values to 0
X.data[X.data < 0] = 0

selector = SelectKBest(score_func=chi2, k=2)
X_new = selector.fit_transform(X, y)
print(X_new)
```
