```python
#1.
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

X = [[1, 2], [3, 4], [5, 6]]
y = [1, 0, 1]

rfe = RFE(estimator=LogisticRegression(), n_features_to_select=1)
X_new = rfe.fit_transform(X, y)
print(X_new)

# adding additional samples improve efficiency
```

```python
#2.
from sklearn.feature_selection import VarianceThreshold

X = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]
selector = VarianceThreshold(threshold=0.1)

# X has only 0's and hence no variance

X = [[0, 1, 0], [0, 2, 1], [1, 1, 0]]
X_new = selector.fit_transform(X)
print(X_new)
```

```python
# 3.
from sklearn.metrics.pairwise import cosine_similarity

# data = [[5, 0, 3], [0, 0, 0]]
# similarity = cosine_similarity(data)

# data[1] contains all 0's hence undefined fr zero vectors
data = [[5, 0, 3], [0, 1, 2]]
similarity = cosine_similarity(data)
print(similarity)
```

```python
# 4.
from sklearn.feature_selection import SelectFromModel
from sklearn.tree import DecisionTreeClassifier

X = [[1, 2], [3, 4], [5, 6]]
y = [1, 0, 1]
selector = SelectFromModel(DecisionTreeClassifier(), threshold="mean")
X_new = selector.fit_transform(X, y)
print(X_new)

## adding additional samples improve efficiency

X1 = [[1, 2], [3, 4], [5, 6], [7, 8], [2, 3]]
y1 = [1, 0, 1, 0, 1]
X_new1 = selector.fit_transform(X1, y1)
print("X_new1: \n", X_new1)
```

```python
#9.
from sklearn.neighbors import NearestNeighbors
import numpy as np

X = np.array([[1, 0, 1], [0, 1, 0], [1, 1, 1]])
model = NearestNeighbors()
model.fit(X)
distances, indices = model.kneighbors([X[0]], n_neighbors=2)
print(indices)
```

```python

#10.
from sklearn.neighbors import NearestNeighbors
import numpy as np

X = np.array([[1, 0, 1], [0, 1, 0], [1, 1, 1]])
model = NearestNeighbors(metric='hamming')
model.fit(X)
distances, indices = model.kneighbors([X[0]], n_neighbors=2)
print("Neighbors (indices):", indices)
```
